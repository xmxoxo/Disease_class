```python
import pandas as pd
```

```python
train_df = pd.read_excel('data_train.xlsx')
train_df.head()
```

```python
train_df['title'].apply(str).head(10)
```

```python
train_df['content'] = train_df['age'].apply(str) + ';' + train_df['diseaseName'].apply(str) + ';' + train_df['conditionDesc'].apply(str) + ';' + train_df['title'].apply(str) + ';' + train_df['hopeHelp'].apply(str)
```

```python
train_df['content'].head(10)
```

```python
con_li = train_df['content'].to_list()
con_len = []
for con in con_li:
    con_len.append(len(str(con)))
con_len
```

```python
train_df['len'] = con_len
```

```python
train_df['label_j'] = train_df['label_j'].values+1
```

```python
train_df.head(10)
```

```python
max(con_len)
```

```python
min(con_len)
```

```python
df = train_df.iloc[:,-4:]
df.head()
```

```python
df.to_csv('data_train.csv', index=None)
```
